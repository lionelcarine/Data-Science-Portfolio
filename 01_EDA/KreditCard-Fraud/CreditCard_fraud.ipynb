# Zielvariable untersuchen ('Class')

```python
df['Class'].value_counts(normalize=True) 
```

```python
sns.countplot(x='Class',data=df)
plt.title("Verteilung: Betrug (1) vs Normal (0)")
plt.show()
# Cette visualisation montre parfaitement le déséquilibre du dataset. 
```

```python
sns.histplot(df['Amount'], bins=100, kde = True)
plt.title("Distributions of Amount")
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Boxplot of Amount by Class")
plt.show()
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
sns.histplot(data=df,x='Hour', hue='Class',bins=24, multiple='stack')
plt.title("Distributions of Hour")
plt.xlabel("Hour(0-23)")
plt.show()
            
```

# Korrelation analysieren

```python
correlations = df.corr()['Class'].sort_values(ascending=False)
print(correlations)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap ='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
important_vars = ['V17', 'V14', 'V12', 'V11','V4','V2'] 
for var in important_vars:
    plt.figure(figsize=(6,4))
    sns.kdeplot(data=df[df['Class']==0], x=var, label="Normal", fill=True)
    sns.kdeplot(data=df[df['Class']==1], x=var, label="Fraude", fill=True)
    plt.title(f"Distribution of {var} by class")
    plt.legend()
    plt.show()
```
