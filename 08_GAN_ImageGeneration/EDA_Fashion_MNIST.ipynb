```python
# Installer numpy si n√©cessaire
import sys
!{sys.executable} -m pip install --user numpy matplotlib seaborn tensorflow
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import fashion_mnist

# Charger les donn√©es
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Nom des classes
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

print(f"Train shape: {x_train.shape}, Test shape: {x_test.shape}")
```

```python
# Exemple d‚Äôimages
def plot_samples(images, labels, class_names, n=10):
    plt.figure(figsize=(15, 3))
    for i in range(n):
        plt.subplot(1, n, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(class_names[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

plot_samples(x_train, y_train, class_names)
```

```python
# R√©partition des classes
sns.countplot(x=y_train)
plt.title("Distribution des classes (Train)")
plt.xticks(ticks=np.arange(10), labels=class_names, rotation=45)
plt.show()
```

```python
# Statistiques des pixels
print("Pixel min:", np.min(x_train), " - max:", np.max(x_train))
print("Moyenne:", np.mean(x_train), " - √âcart-type:", np.std(x_train))
```
